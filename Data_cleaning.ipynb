```python
def process_series(series):
    """
    Convert a pandas Series of dictionary-like strings to a DataFrame with flattened columns.

    Args:
        series (pd.Series): A pandas Series containing dictionary-like strings.

    Returns:
        pd.DataFrame: A DataFrame where each row represents a flattened dictionary.
    """
    processed_list = []
    for element in series:
        # Convert the string to a Python dictionary
        data_dict = ast.literal_eval(element)
        # Flatten the nested dictionary
        nested_dict = data_dict.pop('trendingText', {})
        data_dict.update({key: value for key, value in nested_dict.items()})
        processed_list.append(data_dict)
    # Convert the list of dictionaries into a DataFrame
    return_df=pd.DataFrame(processed_list)
    print(f'shape  ====>  {return_df.shape}')
    print(f'columns  ====>  {return_df.columns}')
    return return_df
# Process the series
detail_df = process_series(df['new_car_detail'])
```

```python
def price_clean(series):
    """
    Cleans and processes a series of price strings into a standardized numeric format.

    This function takes a series of price values in string format (e.g., "10 Lakh", "1.5 Crore", "5000")
    and converts them into a numeric format, scaling values appropriately:
    - "Lakh" is converted to its equivalent in units (multiplied by 100,000).
    - "Crore" is converted to its equivalent in units (multiplied by 10,000,000).
    - Plain numeric strings are converted directly to floats.
    - Invalid or unprocessable entries are replaced with `None`.

    Args:
        series (iterable): An iterable (e.g., list, pandas Series) containing price values as strings.

    Returns:
        pd.DataFrame: A pandas DataFrame with a single column named "price", containing the processed numeric values.

    Example:
        >>> data = ["10 Lakh", "1.5 Crore", "5000", "invalid"]
        >>> price_clean(data)
            price
        0  1000000.0
        1  15000000.0
        2     5000.0
        3        NaN
    """
    processed = []
    for i in series:
        try:
            if 'Lakh' in i:
                replaced = i.replace('Lakh', '')
                striped = replaced.strip()
                processed.append(float(striped) * 100000)
            elif 'Crore' in i:
                replaced = i.replace('Crore', '')
                striped = replaced.strip()
                processed.append(float(striped) * 10000000)
            else:
                processed.append(float(i.strip()))
        except ValueError:
            processed.append(None)
    return pd.DataFrame(processed, columns=["price"])
```

```python
def feature_clean(series):

    """
    Convert a pandas Series of dictionary-like strings to a DataFrame with flattened columns.

    Args:
        series (pd.Series): A pandas Series containing dictionary-like strings.

    Returns:
        pd.DataFrame: A DataFrame where each row represents a flattened dictionary.
    """

    lst=[]
    for element in series:
        sample=ast.literal_eval(element)
        for i in sample['top']:
            lst.append(i['value'])
        for j in sample['data']:
            for k in j['list']:
                lst.append(k['value'])
    unique = sorted(list(set(lst)))
    return unique
unique_feature_list=feature_clean(df['new_car_feature'])
len(unique_feature_list)
```

```python
def feature_assign(unique_features_df):
    """
    Assigns binary features to the input DataFrame based on values extracted
    from the 'new_car_feature' column.

    This function processes the JSON-like string data in the 'new_car_feature'
    column of the input DataFrame. It evaluates the JSON structure using `ast.literal_eval`
    and extracts feature values from the 'top' and 'data' keys. For each unique feature,
    a new column is created in the DataFrame with binary values (1 for presence, 0 otherwise).

    Parameters:
    -----------
    unique_features_df : pd.DataFrame
        A pandas DataFrame containing a 'new_car_feature' column with JSON-like
        string data describing car features.

    Returns:
    --------
    pd.DataFrame
        The modified DataFrame with new columns for each unique feature, where
        the values are binary (1 if the feature is present for a row, 0 otherwise).

    Notes:
    ------
    - The 'new_car_feature' column is expected to contain JSON-like strings with
      specific keys ('top' and 'data') holding lists of feature information.
    - The function modifies the input DataFrame in place by adding new feature columns.
    - Ensure that the input DataFrame is properly formatted before using this function.

    Example:
    --------
    >>> df = pd.DataFrame({'new_car_feature': [
            "{'top': [{'value': 'Feature1'}], 'data': [{'list': [{'value': 'Feature2'}]}]}"
        ]})
    >>> updated_df = feature_assign(df)
    >>> print(updated_df)
       Feature1  Feature2
    0         1         1
    """
    data_frame = unique_features_df
    idx=0
    for element in data_frame['new_car_feature']:
        lst = []
        sample = ast.literal_eval(element)
        for i in sample['top']:
            lst.append(i['value'])
        for j in sample['data']:
            for k in j['list']:
                lst.append(k['value'])
        for e in lst:
            data_frame[e].iloc[idx] = 1
        idx += 1
    return data_frame
```

```python
def specs_clean(series):
    """
    Convert a pandas Series of dictionary-like strings to a DataFrame with flattened columns.

    Args:
        series (pd.Series): A pandas Series containing dictionary-like strings.

    Returns:
        pd.DataFrame: A DataFrame where each row represents a flattened dictionary.
    """
    processed_list = []
    for element in series:
        # convert str into python dict
        single_specs=ast.literal_eval(element)
        single_dict={}
        # Flatten the nested dictionary
        for i in single_specs:
            if i in 'top':
                for top in single_specs[i]:
                    single_dict.update({top['key']:top['value']})
            elif i in 'data':
                for data in single_specs[i]:
                    for j in data['list']:
                        single_dict.update({j['key']:j['value']})
            else:
                pass
        processed_list.append(single_dict)
    # Convert the list of dictionaries into a DataFrame
    return_df=pd.DataFrame(processed_list)
    print(f'shape  ====>  {return_df.shape}')
    print(f'columns  ====>  {return_df.columns}')
    return return_df
specs_df=specs_clean(df['new_car_specs'])
specs_df
```
